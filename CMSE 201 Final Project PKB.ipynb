<font size=3>   Now that we have established that these top emitting countries are a problem, we can assess the ***TOTAL*** about of CO<sub>2</sub> released in the world. This requires using defined functions to obtain the CO<sub>2</sub> levels of all of the countries in a particular year and summing them to recieve the total amount of CO<sub>2</sub> released on Earth in that year. I will be explained what is done each step of the way, to help understand what is being achieved.

````python
#Totalidentify, the defined function that will be used to find the CO2 levels for the World, which is NOT given in 
#the csv file explicitly

Countryspacing = np.arange(271,59621,271)

#This variable is tedious but necessary. The countries same-year values are spaced 271 rows apart, equally. Essentially
#Afghnasitan's 1950 CO2 levels must be added with Albania's 1950 CO2 levels and every other countries 1950 CO2 levels
#to get the TOTAL CO2 levels in 1950. These values are all 271 rows apart on the csv file, hence the 271

def totalidentify(a):
    CO2total=[]
    
    #Just adding an empty list for future reference
    
    for b in range(70):
        total=[]
        
        #This function requires a nested loop for maximum efficiency
        
        for i in range(220):
            
            #For this function, we will start by putting a range of 220 as there are 220 different country values
            #(Some are not countries, and territories or islands, making it greater than the 196 countries there are)
            
            x=int(Countryspacing[i])
            y=CO2data.iloc[x-a,4]
            
            #While this is a bit difficult to understand, this essentially grabs the 2020 values and subtracts the a
            #Input in the defined function. This will get 2020-a, so using a=71 will get the 1950 data
            
            total.append(y)
            
            #Appending to the empty list from the beginning
            
        a-=1
        
        #Now back to the b loop, THIS will repeat the above loop 70 times, for 1950, 1951... 2020 world data
        #To get different values each time, the a value will decrease to get the values for the later years
        #So when this loop repeats, the a value will be one less and retrieve the 1951 values
        
        totsum=sum(total)
        
        #Totsum is the variable which contains the explicit numerical value of the total CO2 in the year. For the first
        #loop, the value is 230184741287. This means every 1950 value for every country has been added and is set to
        #this variable. This value will go into the CO2total list for the final
        
        CO2total.append(totsum)
    return CO2total

#This data was intended to have 71 as the variable. It works with other values but 91, for example, will give values
#From 1930 to 2000. This is not quite what I want so 71 as the variable gives 1950 to 2020

totalidentify(71)
````

````python
#Now, plot this graph for a visual representation of the CO2 levels in the world

#Set up the graph

sns.set_style('whitegrid')
figure(figsize=(8,5), dpi=100)

Years = np.arange(1950,2020, 1)
total70years = totalidentify(71)

#Basic plotting, showing the data that was recieved above

plt.plot(Years, total70years, color = 'g', linewidth = 4)
plt.title('Overall $CO_{2}$ levels in the Atmosphere', fontsize = 18)
plt.ylabel('$CO_{2}$ Emission (Tons)', fontsize = 14)
plt.xlabel('Year', fontsize = 14)
````

<font size=3> Now that the world data has been found, we can introduce a level of dangerous CO<sub>2</sub> that is realistically the tipping point of Earth and its habitability. According to the Yale article introduced above, the realistic limit that we have on the level of CO<sub>2</sub> that is allowed is 500ppm+. This is because at this level of CO<sub>2</sub>, the sea levels will rise drastically due to ice caps melting and agriculture will become much more difficult as violent weather conditions will ruin more crops. The average temperature will raise by more than 2<sup>o</sup>C and this will make some countries so hot to the point of inhabitability. This is roughly translating to 3.5 trillion tons of CO<sub>2</sub> and while that may seem like a lot more than we have; at this pace we don't even have 50 years left.

````python
#First, let's add a line at 3.5 trillion tons to represent the limit and the current graph of the World CO2

figure(figsize=(8,5.5), dpi=100)
sns.set_style('whitegrid')

#Also adding a thin shading above to indicate very dangerous levels of CO2

years=np.linspace(1951,2020,70)
x_fill = np.linspace(1951, 2020, 10000)
yline = np.linspace(3.5E12,3.5E12, 70)

#Now plotting all of the important information

plt.fill_between(x_fill, x_fill+3.5E12, y2=5E12, alpha=0.25, color='c', label='Very unsafe $CO_{2}$ Levels')
plt.plot(years,totalidentify(71), label='CO2 Worldwide', linewidth=3)
plt.plot(years, yline, '--', label='Limit of safe $CO_{2}$ Levels', color='c', linewidth=3)
plt.title('World $CO_{2}$ Trend and Danger Zone', fontsize=15)
plt.ylabel('$CO_{2}$ Levels in Tons', fontsize =13)
plt.xlabel('Year', fontsize=13)
plt.legend()
````

````python
#Now that we have a graph with the dangerous levels of CO2 along with the datapoints, We now whave to predict future
#points using a complex equation. I will manually retreive the equation online as it is a helpful resource

#Now, creating a function to find the list of future points

def futurepoints(i):
    
    #The i input represents the number of years past 1950, ex: i=70 would just be 1950-2020 points
    
    futurepoints=[]
    
    #Create a list for future use
    
    for x in range(i):
        
        #This simply takes the log function of the current points obtained and inserts the x values
        
        future=2.1316*(10**8)*(x**2)+5.15771*(10**9)*x-1.10591*(10**10)+2.5*10**11
        futurepoints.append(future)
        
        #Append each point to the empty list
        #Repeat i number of times
        
    return futurepoints

#This is now the function that will be used for future points and the line of best fit essentially
#This gives an extremely close similarity to the graph, with a less than 0.1% error, so error bars will be negligible
````

````python
figure(figsize=(9,6), dpi=100)
yline2=np.linspace(3.5E12,3.5E12,150)
yearsfuture=np.linspace(1951,2100,150)
x_fill2 = np.linspace(1951, 2100, 10000)

#Repeat the same graph but add an intersection point and the future points

plt.fill_between(x_fill2, x_fill2+3.5E12, y2=6E12, alpha=0.25, color='c', label='Very unsafe $CO_{2}$ Levels')
plt.plot(yearsfuture,yline2,'c-.', linewidth=2.5)
plt.plot(yearsfuture,futurepoints(150), label='Estimated $CO_{2}$ Levels', color='b', linewidth=2)
plt.scatter(2063, 3.5E12, label='Point of Realistic Irreversibility', linewidth=2, color='k')
plt.title('Future $CO_{2}$ Levels and Danger Zone Intersection', fontsize=16)
plt.ylabel('$CO_{2}$ Levels (Tons)', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.legend()

#This gives a clean graph that is easily digestible and accurate
````

<font size=3> For fun, I will add a scenario in let's say, 2025, we decide to change our wrongdoings and fix the damages that us and our ancestors have commited to our beloved planet today. This will require changing the future point values and defining a new function for it. The limit of CO<sub>2</sub> will still be acceptable and will continue to be used as a reference. 
    \
    \
In this scenario, let's say, the world's major powers have decided to make make heavy fossil fuel use punishable, and enforcing corporations to replace 50% of all current energy sources using fossil fuels and replace it with a renewable equivalent, such as _solar panels_ or _wind turbines_. 
    

````python
#I will now create a new defined function, using the initial defined function to start and then ending with a lower curve

#The new function will be named optfuturepoints for optimistism

def optfuturepoints(i):
    
    #The i input represents the number of years past 1950, ex: i=70 would just be 1950-2020 points
    
    optfuturepoints=[]
    
    #Create a list for future use
    
    for j in range(i):
        
        #Only difference is adding an elif statement, if the input is less than 75 (1950-2024), then the graph will 
        #Remain the same but after, during 2025 and beoynd, the new function is applied
        
        if j < 75:
            optfuture=futurepoints(i)[j]
        else:
            optfuture=-65922000*(j-80)**2+24259100000*j
        optfuturepoints.append(optfuture)
        
        #Append each point to the empty list
        #Repeat i number of times
        
    return optfuturepoints

#This is now the function that will be used for optimistic future points
````

````python
figure(figsize=(9,6), dpi=100)
yline2=np.linspace(3.5E12,3.5E12,150)
yearsfuture=np.linspace(1951,2100,150)
x_fill2 = np.linspace(1951, 2100, 10000)

#Repeat the same graph but add an intersection point and the future points

plt.fill_between(x_fill2, x_fill2+3.5E12, y2=6E12, alpha=0.25, color='c', label='Very unsafe $CO_{2}$ Levels')
plt.plot(yearsfuture,yline2,'c-.', linewidth=2.5)
plt.plot(yearsfuture,optfuturepoints(150), label='Estimated $CO_{2}$ Levels', color='b', linewidth=2)
plt.axvline(2063, label='Initial line of Irreversibility', linewidth=2, color='k', alpha=0.3)
plt.title('Optimistic Future $CO_{2}$ Levels and Danger Zone', fontsize=16)
plt.ylabel('$CO_{2}$ Levels (Tons)', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.legend()

#This gives a clean graph that is easily digestible and accurate
````
